```python
# Import Library

import librosa
import tensorflow as tf
import random
import os
import numpy as np
import librosa.display
import matplotlib.pyplot as plt 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
# Necessary Variabel

SAMPLE_RATE = 16_000
VALIDATION_SPLIT = 0.1
SHUFFLE_SEED = 42
FFT_SIZE = 1024
HOP_SIZE = 128
N_MELS = 128
MFCC_BINS = 128
MAX_LEN_PAD = 80
WIDTH_IMAGE = 300
LENGTH_IMAGE = 300
CHANNELS = 3

# Directory Path

dataset_directory = os.getcwd() + "/AudioClassification Dataset"
dataset_labels = os.listdir(dataset_directory)
```

```python
# Necessary Function

def resample_audio(path):
    audio, sr = librosa.load(path, sr=SAMPLE_RATE)
    if len(audio) < SAMPLE_RATE: 
        audio = np.pad(audio, (0,16000-len(audio)), "constant")
    else:
        audio = audio[:SAMPLE_RATE]
    return audio

def stft(audio):
    audio_stft = librosa.stft(y = audio, n_fft = FFT_SIZE, hop_length = HOP_SIZE, center=True, pad_mode='constant')
    amp = np.abs(audio_stft)**2

    return amp

def mel_frequency(spectrogram):
    mel = librosa.filters.mel(sr=SAMPLE_RATE, n_fft=FFT_SIZE, n_mels=N_MELS)
    
    mel_spec = mel.dot(spectrogram)
    db_mel_spect = librosa.power_to_db(mel_spec, ref=np.max)
    
    return db_mel_spect

def mfcc(mel_freq):
    mfcc = librosa.feature.mfcc(S=mel_freq, sr=SAMPLE_RATE, n_mfcc=MFCC_BINS, n_fft=FFT_SIZE, hop_length=HOP_SIZE, n_mels=N_MELS)
    
    if (MAX_LEN_PAD > mfcc.shape[1]):
        pad = MAX_LEN_PAD - mfcc.shape[1]
        mfcc = np.pad(mfcc, pad_width=((0, 0), (0, pad)))
    else:
        mfcc = mfcc[:, :MAX_LEN_PAD]
    return mfcc

def delta_mfcc(mfcc):
    delta_mfcc = librosa.feature.delta(mfcc)
    
    return delta_mfcc

def delta_delta_mfcc(mfcc):
    delta_delta_mfcc = librosa.feature.delta(mfcc, order=2)
    
    return delta_delta_mfcc
```

```python
# Augmentation Function

def white_noise(audio):
    noise_factor = 0.05
    white_noise = np.random.randn(len(audio)) * noise_factor
    white_noise_audio = audio + white_noise
    return white_noise_audio

def change_speed(audio):
    rate = 0.6
    change_speed_audio = librosa.effects.time_stretch(audio, rate = rate)
    return change_speed_audio

def change_pitch(audio):
    pitch_factor = 4
    change_pitch_audio = librosa.effects.pitch_shift(audio, SAMPLE_RATE, pitch_factor)
    return change_pitch_audio
```

```python
# Mel-Spectrogram Augmentation Function
def time_mask(spec, time_masking_max_percentage=0.15, num_masks=1):
    spec_aug = spec.copy()
    all_frames_num = spec_aug.shape[0]
    for i in range(num_masks):
        time_percentage = random.uniform(0.0, time_masking_max_percentage)
        
        num_frames_to_mask = int(time_percentage * all_frames_num)
        t0 = np.random.uniform(low=0.0, high=all_frames_num - num_frames_to_mask)
        t0 = int(t0)
        spec_aug[t0:t0 + num_frames_to_mask, :] = 0.000
    return spec_aug

def freq_mask(spec, freq_masking_max_percentage=0.10, num_masks=1):
    spec_aug = spec.copy()
    all_freqs_num = spec_aug.shape[1]
    for i in range(num_masks):        
        freq_percentage = random.uniform(0, freq_masking_max_percentage)
        
        num_freqs_to_mask = int(freq_percentage * all_freqs_num)
        f0 = np.random.uniform(low=0.0, high=all_freqs_num - num_freqs_to_mask)
        f0 = int(f0)
        spec_aug[:, f0:f0 + num_freqs_to_mask] = 0.000
    return spec_aug
```

```python
audio_paths = []
labels = []

for label, name in enumerate(dataset_labels):
    
    dir_path = os.path.join(dataset_directory, name)
    speaker_sample_paths = [os.path.join(dir_path, filepath) for filepath in os.listdir(dir_path) if filepath.endswith(".wav")]
    
    audio_paths += speaker_sample_paths
    labels += [label] * len(speaker_sample_paths)
    
```

```python
# Resample Audio
resampled_audio = [resample_audio(path) for path in audio_paths]
```

```python
# Audio Augmentation
white_noise_audio = [white_noise(audio) for audio in resampled_audio]
pitch_audio = [change_pitch(audio) for audio in resampled_audio]
speed_audio = [change_speed(audio) for audio in resampled_audio]
```

```python
# Combine Augmented Audio with Resampled Audio
combined_audio = resampled_audio + white_noise_audio + pitch_audio + speed_audio
labels_combined = labels * 4
```

```python
# Convert to Frequency Domain (STFT) Spectrogram
audio_spectrogram = [stft(audio) for audio in combined_audio]
```

```python
# Spectrogram
for i in range(3):
    plt.subplot(3, 1, i+1)
    librosa.display.specshow(librosa.amplitude_to_db(audio_spectrogram[i]**(1/2), ref=np.max), sr=SAMPLE_RATE, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    plt.show()
```

```python
# Convert Spectrogram Into Mel Spectrogram|
audio_mel_spectrogram = [mel_frequency(spectrogram) for spectrogram in audio_spectrogram]
```

```python
# Mel Frequency Spectrogram
for i in range(3):
    plt.subplot(3, 1, i+1)
    librosa.display.specshow(audio_mel_spectrogram[i], y_axis='mel', fmax=8000, x_axis='time', sr=SAMPLE_RATE)
    plt.title('Mel Spectrogram')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
```

```python
# Augmented Mel-Spectrogram
audio_mel_augmented = [freq_mask(time_mask(mel)) for mel in audio_mel_spectrogram]
audio_mel_combined = audio_mel_augmented + audio_mel_spectrogram
labels_combined_mel = labels_combined * 2
```

```python
# Split Train and Validation
x_train, x_val, y_train, y_val = train_test_split(cnn_audio, cnn_labels, stratify=cnn_labels, test_size=VALIDATION_SPLIT)
x_val, x_test, y_val, y_test = train_test_split(x_val, y_val, stratify=y_val, test_size=0.2)
```

```python
# Model CNN

inputs = tf.keras.layers.Input(shape=(MFCC_BINS, MAX_LEN_PAD, CHANNELS))
    
x = tf.keras.layers.Conv2D(16, (3, 3), activation = 'relu', padding='same')(inputs)
x = tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu', padding='same')(x)
x = tf.keras.layers.MaxPooling2D((2,2), padding='same')(x)
x = tf.keras.layers.BatchNormalization()(x)

for i in range(2, 6):
    x = tf.keras.layers.Conv2D(i * 16, (3, 3), activation = 'relu', padding='same')(x)
    x = tf.keras.layers.Conv2D(i * 32, (3, 3), activation = 'relu', padding='same')(x)
    x = tf.keras.layers.MaxPooling2D((2,2), padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dropout(0.3)(x)


x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)

outputs = tf.keras.layers.Dense(len(dataset_labels), activation='softmax')(x)
model = tf.keras.models.Model(inputs=inputs, outputs=outputs)

```
